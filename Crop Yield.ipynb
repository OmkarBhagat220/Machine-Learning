Exploratory Data Analysis

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
#Remove countries with less than 100 record
country_counts = df['Area'].value_counts()
countries_to_drop = country_counts[country_counts<100].index.tolist()
df_filter = df[~df['Area'].isin(countries_to_drop)]
df = df_filter.reset_index(drop=True)
```

```python
df
```

```python
d_f = df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
cat = d_f.select_dtypes(include=['object']).columns.tolist()
label_encoder = LabelEncoder()
for column in cat:
    d_f[column] = label_encoder.fit_transform(d_f[column])
```

```python
sns.heatmap(d_f.corr(), annot = True, cmap= 'PuOr')
```

Store Feature Matrix in X and Respose (target) in vector y

Sepreate Data and Label

Sepreate Features and Target

```python
X, y = d_f.drop(labels='hg/ha_yield', axis=1), d_f['hg/ha_yield']
```

Model building and evaluation /
Test and Train Data

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=1)
```

RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(Xtrain, ytrain)
ypred = rfr.predict(Xtest)
rfr_train=rfr.score(Xtrain, ytrain)
rfr_test=rfr.score(Xtest, ytest)

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
rfr_MSE = mean_squared_error(ytest,ypred)
rfr_R2_score = r2_score(ytest,ypred)
rfr_MAE = mean_absolute_error(ytest,ypred)
rfr_MAPE = mean_absolute_percentage_error(ytest, ypred)

print(f'The accuracy of the Random Forest Regressor Model Train is {rfr_train}')
print(f'The accuracy of the Random Forest Regressor Model Test is {rfr_test}')
print(f'The MSE of the Random Forest Regressor is {rfr_MSE}')
print(f'The r2score of the Random Forest Regressor is {rfr_R2_score}')
print(f'The MAE of the Random Forest Regressor is {rfr_MAE}')
print(f'The MAPE of the Random Forest Regressor is {rfr_MAPE}')


plt.scatter(ytest, ypred, s=10, color= 'red')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.plot([min(ytest), max(ytest)],[min(ytest), max(ytest)],color = 'green', linewidth=4)
plt.show()
```

DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor()
DTR.fit(Xtrain, ytrain)
ypred = DTR.predict(Xtest)
DTR_train=DTR.score(Xtrain, ytrain)
DTR_test=DTR.score(Xtest, ytest)

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
DTR_MSE = mean_squared_error(ytest,ypred)
DTR_R2_score = r2_score(ytest,ypred)
DTR_MAE = mean_absolute_error(ytest,ypred)
DTR_MAPE = mean_absolute_percentage_error(ytest, ypred)

print(f'The accuracy of the Decision Tree Model Train is {DTR_train}')
print(f'The accuracy of the Decision Tree Model Test is {DTR_test}')
print(f'The MSE of the Decision Tree is {DTR_MSE}')
print(f'The r2score of the Decision Tree is {DTR_R2_score}')
print(f'The MAE of the Decision Tree is {DTR_MAE}')
print(f'The MAPE of the Decision Tree is {DTR_MAPE}')


plt.scatter(ytest, ypred, s=10, color= 'red')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.plot([min(ytest), max(ytest)],[min(ytest), max(ytest)],color = 'green', linewidth=4)
plt.show()
```

```python
results = []
results.append({'Model': 'Decision Tree', 'Accuracy': DTR_test, 'Mean Squared Error': DTR_MSE, 'R2_score': DTR_R2_score, 'Mean Absolute Error':DTR_MAE, 'Mean Absolute Percentage Error':DTR_MAPE})
results.append({'Model': 'Random Forest', 'Accuracy': rfr_test, 'Mean Squared Error': rfr_MSE, 'R2_score': rfr_R2_score, 'Mean Absolute Error':rfr_MAE, 'Mean Absolute Percentage Error':rfr_MAPE})
dff = pd.DataFrame(results, columns=['Model', 'Accuracy', 'Mean Squared Error', 'R2_score','Mean Absolute Error','Mean Absolute Percentage Error'])
dff
```
